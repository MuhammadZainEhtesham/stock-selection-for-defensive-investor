```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
#getting all the listed companies
url = 'https://dps.psx.com.pk/market-watch'
res = requests.get(url)
dfs = pd.read_html(res.text)
df = dfs[0]
companies = list(df['SYMBOL'])
```

```python
df_criterion_passed = pd.DataFrame(columns = ['SYMBOL','PE Ratio'])
df_criterion_failed = pd.DataFrame(columns = ['SYMBOL','PE Ratio'])
for company in companies:
    try:
        url = os.path.join('https://dps.psx.com.pk/company/',company)
        response = requests.get(url)
        text = response.text
        soup = BeautifulSoup(text,'html.parser')
        lst = soup.find_all('div',class_ = 'stats_item')
        t = lst[14]
        t = str(t)
        t = t.split('>')[-3].split('<')[0]
        pe_ratio = float(t)
        if pe_ratio<15:
            df_criterion_passed = df_criterion_passed.append({'SYMBOL':company,'PE Ratio':pe_ratio}, ignore_index = True)
            print(company, ' analyzed')
        else:
            df_criterion_failed = df_criterion_failed.append({'SYMBOL':company,'PE Ratio':pe_ratio},ignore_index = True)
            print(company, ' analyzed')
    except:
        print('error with ', company)
    finally:
        continue
```
